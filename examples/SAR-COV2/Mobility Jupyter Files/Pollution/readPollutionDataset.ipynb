## Merge Tables

```python
quality = pd.read_csv('dataset.csv')
print(len(quality))
quality = quality.rename(columns={'CODI_PROVINCIA':'Codi_provincia',
                        'PROVINCIA':'Provincia',
                       'CODI_MUNICIPI':'Codi_municipi',
                       'MUNICIPI':'Municipi',
                       'ESTACIO':'Estacio',
                       'CODI_CONTAMINANT':'Codi_Contaminant',
                       'ANY':'Any',
                       'MES':'Mes',
                       'DIA':'Dia'})
quality.head(10)
```

```python
stations = pd.read_csv('qualitat_aire_estacions.csv')
stations = stations.drop(columns = 'Codi_Contaminant')
stations = stations.drop_duplicates()
print(len(stations))
stations.sort_values('Estacio')
```

```python
contaminants = pd.read_csv('qualitat_aire_contaminants.csv')
print(len(contaminants))
contaminants.head(1)
```

```python
quality_stations = pd.merge(quality, stations, on=['Estacio'])
len(quality_stations)
```

```python
quality_stations_contaminant = pd.merge(quality_stations, contaminants, on = ['Codi_Contaminant'])
len(quality_stations_contaminant)
```

```python
quality_stations_contaminant.to_csv('qualitat_estacions_contaminants.csv')
```
